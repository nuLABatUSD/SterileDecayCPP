```python
import numpy as np
import matplotlib.pyplot as plt
import ODESolve_slow as ODE
#import ODESolve as ODE

########################
#
# ODESolve solves the ODE:
# dy/dx = f(x, y, p)
# 
# y is an array 
# f is an array of the derivatives with length len(y)
# p is an array of parameters used to define the ODEs and provide flexibility
#
# The function f(x, y, p) must be defined 
```

```python
y0 = np.ones(3)
p = np.zeros(4)

p[0] = 0
p[1:] = np.random.rand(3)

x0 = 0
dx0 = 0.01
```

```python
results = ODE.ODEOneRun(x0, y0, dx0, p, 100, 2, 20)
```

```python
plt.figure()
for i in range(3):
    plt.plot(results[0], results[1][:,i], 'o-',markersize=3,label=r"$\lambda = -{}$".format(np.round(p[i+1], 3)))
plt.legend(loc='upper right')
plt.title(r"Solving $y' = - \lambda y^2$ with $y(0) = 1$")
plt.show()
```
